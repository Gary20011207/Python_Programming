1. 常態分佈的累積密度函數

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

loc1, scale1 = 0, 1
x1 = np.linspace(stats.norm.ppf(0.001, loc1, scale1), stats.norm.ppf(0.999, loc1, scale1), 1000)
plt.plot(x1, stats.norm.cdf(x1), "r-", label = "norm1 cdf")
loc2, scale2 = 2, 1
x2 = np.linspace(stats.norm.ppf(0.001, loc2, scale2), stats.norm.ppf(0.999, loc2, scale2), 1000)
plt.plot(x2, stats.norm.cdf(x2, loc2, scale2), "g--", label = "norm2 cdf")
loc3, scale3 = 0, 2
x3 = np.linspace(stats.norm.ppf(0.001, loc3, scale3), stats.norm.ppf(0.999, loc3, scale3), 1000)
plt.plot(x3, stats.norm.cdf(x3, loc3, scale3), "b:", label = "norm3 cdf")
plt.legend()
plt.show()
```

2. 函式的根

```python
import numpy as np
from scipy.optimize import root

def f(x):
    return (x + np.cos(x))

sol = root(f, 0)
print(sol.x)
```

3. 解方程式

```python
import numpy as np
from scipy.optimize import root

def fun(x):
    return (x[0] * np.cos(x[1]) - 4, x[1] * x[0] - x[1] - 5)

sol = root(fun, [1, 1])
print(sol.x)
```

4. 函數的最小值

```python
from scipy.optimize import minimize_scalar

def f1(x):
    return (3 * x ** 4 - x ** 2 + 5)

def f2(x):
    return (x * (x - 2) * ((x + 2) ** 2))

res1 = minimize_scalar(f1)
res2 = minimize_scalar(f2)
print("當x為{0}時，函數f1(x)有最小值為{1}。".format(res1.x, f1(res1.x)))
print("當x為{0}時，函數f2(x)有最小值為{1}。".format(res2.x, f2(res2.x)))
```

5. 插值

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
f = lambda x: (x - 1) * (x - 2) * (x - 3)
x = np.linspace(0, 5, num = 11)
y = f(x)

g1 = interp1d(x, y)
g2 = interp1d(x, y, kind = "quadratic")

xnew = np.linspace(0, 5, num = 121)
ynew1 = g1(xnew)
ynew2 = g2(xnew)

plt.plot(x, y, "o", label = "data")
plt.plot(xnew, ynew1, "-", label = "linear")
plt.plot(xnew, ynew2, "--", label = "quadratic")
plt.legend()
plt.show()

#亦可寫成
#plt.plot(x, y, "o", xnew, ynew1, "-", xnew, ynew2, "--")
#plt.legend(["data", "linear", "quadratic"])
#plt.show()
```
