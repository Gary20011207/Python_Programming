1. 繪製數學函式

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 10000)
y1 = 20 * np.sin(x)
y2 = x * x * np.cos(x) + 0.5
plt.plot(x, y1, "b-")
plt.plot(x, y2, "r--")
plt.show()
```

2. 牧場動物飼養比例

```python
import matplotlib.pyplot as plt

animals = ["cow", "sheep", "duck", "chicken", "others"]
numbers = [10, 15, 28, 12, 7]
colors = ["lightgreen", "lightblue", "yellow", "pink", "silver"]
plt.pie(numbers, labels = animals, colors = colors, shadow = True,
explode = (0.1, 0, 0, 0, 0), autopct = "%1.1f%%")
plt.axis("equal")
plt.show()
```

3. 擲骰子次數統計

```python
import numpy as np
import matplotlib.pyplot as plt

samples = np.random.randint(1, 7, size = 101)
x_range = np.arange(1, 7, 1)
times = np.bincount(samples, weights = None, minlength = 0)[1:7]
tl = ['1', '2', '3', '4', '5', '6']
plt.bar(x = x_range, height = times, tick_label = tl)
plt.show()
```
